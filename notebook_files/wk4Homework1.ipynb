```python
import pandas as pd
from pathlib import Path
```

```python
school_complete = Path("C:/Users/lhs70/OneDrive/Desktop/wk4_resource/schools_complete.csv")
student_complete = Path("C:/Users/lhs70/OneDrive/Desktop/wk4_resource/students_complete.csv")

school_df = pd.read_csv(school_complete)
student_df = pd.read_csv(student_complete)

school_df.head()
```

```python
student_df.head()
```

```python
# Total number of unique schools
total_unique_schools = len(school_df['school_name'].unique())

# Total number of students
total_students = school_df['size'].sum()

# Total budget
total_budget = school_df['budget'].sum()

# Average maths score
average_maths_score = student_df['maths_score'].mean()

# Average reading score
average_reading_score = student_df['reading_score'].mean()

# Percentage passing maths
passing_maths_percentage = (len(student_df[student_df['maths_score'] >= 50]) / len(student_df)) * 100

# Percentage passing reading
passing_reading_percentage = (len(student_df[student_df['reading_score'] >= 50]) / len(student_df)) * 100

# Percentage overall passing (passed both maths and reading)
passed_both = student_df[(student_df['maths_score'] >= 50) & (student_df['reading_score'] >= 50)]
overall_passing_percentage = (len(passed_both) / len(student_df)) * 100

# Creating a summary DataFrame
summary = pd.DataFrame({
    'Total Unique Schools': [total_unique_schools],
    'Total Students': [total_students],
    'Total Budget': [total_budget],
    'Average Maths Score': [average_maths_score],
    'Average Reading Score': [average_reading_score],
    '% Passing Maths': [passing_maths_percentage],
    '% Passing Reading': [passing_reading_percentage],
    '% Overall Passing': [overall_passing_percentage]
})
              
# Displaying the summary DataFrame
print(summary)
```

```python
# Push the remade DataFrame to a new CSV file
summary.to_csv("C:/Users/lhs70/OneDrive/Desktop/wk4_resource/to_submit/lga_summary.csv",
                  encoding="utf-8", index=False, header=True)
```
