```python
import pandas as pd
from pathlib import Path
```

```python
school_summary = Path("C:/Users/lhs70/OneDrive/Desktop/wk4_resource/to submit/schools_summary.csv")
student_complete = Path("C:/Users/lhs70/OneDrive/Desktop/wk4_resource/students_complete.csv")

school_summary_df = pd.read_csv(school_summary, encoding='utf-8')
student_df = pd.read_csv(student_complete, encoding='utf-8')

school_summary_df.head(2)
```

```python
# get top 5 schools
top_schools = school_summary_df[["School name", "Overall Passing (%)"]].sort_values(by="Overall Passing (%)", ascending=False)
top_schools = top_schools.head(5)
top_schools
```

```python
# get bottom 5 schools
bottom_schools = school_summary_df[["School name", "Overall Passing (%)"]].sort_values(by="Overall Passing (%)", ascending=True)
bottom_schools = bottom_schools.head(5)
bottom_schools
```

```python
student_df.head(2)
```

```python
# group-by according to school and year
new_df = student_df[["school_name", "year", "maths_score", "reading_score", "student_name"]]
new_df.head()
```

```python
# Group by 'school_name' and 'year'
grouped_df = new_df.groupby(["school_name", "year"])  
```

```python
maths_by_year_df = grouped_df['maths_score'].mean()
maths_by_year_df
```

```python
reading_by_year_df = grouped_df['reading_score'].mean()
reading_by_year_df
```
