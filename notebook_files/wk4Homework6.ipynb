```python
import pandas as pd
from pathlib import Path
```

```python
school_complete = Path("C:/Users/lhs70/OneDrive/Desktop/wk4_resource/schools_complete.csv")
student_complete = Path("C:/Users/lhs70/OneDrive/Desktop/wk4_resource/students_complete.csv")

school_df = pd.read_csv(school_complete)
student_df = pd.read_csv(student_complete)

school_df.head(2)
```

```python
type_df = school_df[["school_name","type"]]
type_df.head(2)
```

```python
student_df.head(2)
```

```python
school_type_df = pd.merge(student_df, type_df, on="school_name", how='left')
school_type_df.head(2)
```

```python
type_maths_scores = school_type_df.groupby(["type"], observed=False)["maths_score"].mean()
type_reading_scores = school_type_df.groupby(["type"], observed=False)["reading_score"].mean()

type_1 = pd.DataFrame({
    'Average Maths Score': type_maths_scores,
    'Average Reading Score': type_reading_scores
})

type_1
```

```python
schools_summary = Path("C:/Users/lhs70/OneDrive/Desktop/wk4_resource/to_submit/schools_summary.csv")
summary_df = pd.read_csv(schools_summary)

summary_df.head(2)
```

```python
passing_df = summary_df[["School name","Passing Maths (%)", "Passing Reading (%)", "Overall Passing (%)"]]
passing_df.head(2)
```

```python
passing_df = passing_df.copy()
passing_df.rename(columns={'School name': 'school_name'}, inplace=True)
passing_df.head(2)
```

```python
school_type2_df = pd.merge(passing_df, type_df, on="school_name", how='left')
school_type2_df.head(2)
```

```python
type_passing_maths = school_type2_df.groupby(["type"], observed=False)["Passing Maths (%)"].mean()
type_passing_reading = school_type2_df.groupby(["type"], observed=False)["Passing Reading (%)"].mean()
type_passing_overall = school_type2_df.groupby(["type"], observed=False)["Overall Passing (%)"].mean()

type_2 = pd.DataFrame({
    '% passing maths': type_passing_maths,
    '% passing reading': type_passing_reading,
    '% overall passing': type_passing_overall
})

type_2
```

```python
type_summary = pd.merge(type_1, type_2, on="type", how='outer')
type_summary
```

```python
type_summary.to_csv("C:/Users/lhs70/OneDrive/Desktop/wk4_resource/to_submit/type_summary.csv", encoding="utf-8", index=True, header=True)
```
